```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xvals = np.linspace(0,5,1000) # 1000 points between 0 and 5
```

## Radioactive decay
$A = A_0 e^{-\lambda x}$, where $A_0 = 1$ is initial proportion of parent isotope, $\lambda$ is the activity constant and A is the final proportion of parent isotope. The child isotope proportion is got with $1-A$

```python
l = -np.log(1/2) # Scaling constant so that x-axis matches half-lives
azero = 1
parent_curve = list(map(lambda x: azero * np.exp(-l * x), xvals))
child_curve = list(map(lambda x: 1 - azero * np.exp(-l * x), xvals))
```

```python
plt.figure(figsize=(10,6))
plt.plot(xvals, parent_curve, color='r')
plt.plot(xvals, child_curve, color='b')
plt.legend(["Hajoava isotooppi","Hajoamistuote"])
plt.xlabel("Kulunut aika puoliintumisajoissa")
plt.ylabel("Aineen suhteellinen osuus")
plt.yticks([0,0.25,0.5,0.75,1.0])
plt.axvline(x = 1, color = 'g', label = 'axvline - full height', linestyle=':')
plt.axvline(x = 2, color = 'g', linestyle = ':')
```
